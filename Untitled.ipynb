```python
import plotly
import plotly.express as px
import pandas as pd 
import json
import glob
import re
import os
```

```python
# input : output_0.18.json
# output : 0.18
def extract_version(filename):
    # Rechercher la version dans le nom de fichier
    match = re.search(r'output_([0-9.]+)(?=.json$)', filename)
    if match:
        return match.group(1)
    return filename  # Retourner le nom du fichier si la version n'est pas trouvée
```

```python
# rechercher tous les .jon du directory
file_pattern = "output_*.json"
files = glob.glob(file_pattern)
print(files)
```

```python
# trier par valeur de version (pour la regression)
files.sort(key=lambda f: extract_version(f))
print(files)
```

```python
# ouvrir et lire tous les fichiers json
all_benchmarks_df = []
for file_path in files : 
    with open(file_path, "r") as file : 
        json_data = json.load(file)
    # Extraire les benchmarks en tant que DataFrame
    benchmarks_df = pd.json_normalize(json_data["benchmarks"])
    # Ajouter une colonne pour extraire la taille depuis le nom du test
    benchmarks_df["size"] = benchmarks_df["name"].apply(lambda x: int(re.search(r'/([0-9]+)$', x).group(1)) if re.search(r'/([0-9]+)$', x) else None)
    # Ajouter une colonne pour identifier le cas-test sans la taille
    benchmarks_df["test_case"] = benchmarks_df["name"].apply(lambda x: re.sub(r'/[0-9]+$', '', x))
    # ajout du filename
    benchmarks_df["filename"] = os.path.basename(file_path)
    # Ajouter le DataFrame à la liste
    all_benchmarks_df.append(benchmarks_df)        
```
